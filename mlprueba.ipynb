```python
import requests
cat_id = 'MLA3937'
offset = 50
url = f'https://api.mercadolibre.com/sites/MLA/search?category={cat_id}&offset={offset}'
request = requests.get(url)
items = request.json()
results=items['results']
results
```

```python
titles=[]
ids=[]
domain=[]
price=[]
orprice=[]
rep=[]
saprice=[]
solds=[]
available=[]
for k in results:
      #print(k['seller']['seller_reputation']['level_id'])
      titles.append(k['title'])
      ids.append(k['id'])
      domain.append(k['domain_id'])
      price.append(k['price'])
      orprice.append(k['original_price'])
      rep.append(k['seller']['seller_reputation']['level_id'])
      saprice.append(k['sale_price'])
      solds.append(k['sold_quantity'])
      available.append(k['available_quantity'])
      #price.append(k['price'])
      #price.append(k['price'])
      #attributes=[]
      #for l in k["attributes"]:
          #print(k["member_skill_list"]["skill"])
      #    attributes.append(l['id'])
          #print(skills)
      #print(attributes)
      #show=attributes
      #todos son de argentina

len(domain)
df = pd.DataFrame(
    {'Titles': titles,
     'ids': ids,
     'domain': domain,
     'price': price,
     'orprice': orprice,
     'rep': rep,
     'saleprice': saprice,
     'solds': solds,
     'availables': available,})
df.head(100)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Ejemplo de títulos de productos
titulo1 = df['Titles'][0]
titulo2 = df['Titles'][2]

# Crear un vectorizador TF-IDF
vectorizer = TfidfVectorizer()

# Transformar los títulos en vectores TF-IDF
tfidf_matrix = vectorizer.fit_transform([titulo1, titulo2])

# Calcular la similitud del coseno
similitud_coseno = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])

print("Similitud del coseno entre los títulos de los productos:", similitud_coseno[0][0])
```

```python
resultado_filtrado = resultado_filtrado.reset_index(drop=True)
resultado_filtrado.head()
```

```python
vectorizer = TfidfVectorizer()
cosdis=[]
# Transformar los títulos en vectores TF-IDF
for k,l in resultado_filtrado.iterrows():
  #print(k)
  tfidf_matrix = vectorizer.fit_transform([Actualprod['Titles'], resultado_filtrado['Titles'][k]])

  # Calcular la similitud del coseno
  similitud_coseno = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])
  cosdis.append(similitud_coseno[0,0])

print(cosdis)
resultado_filtrado['cosdis']=cosdis
resultado_filtrado.head(10)
```

```python
df_ordenado = resultado_filtrado.sort_values(by='cosdis', ascending=False)
df_ordenado = df_ordenado.reset_index(drop=True)
df_ordenado=df.drop(df.index[0])
best=df_ordenado.iloc[:4]
```
